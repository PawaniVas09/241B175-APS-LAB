```python
# Exercise 1
def func1(a, b): #Time complexicity is O(b)
  result = 0
  for i in range(0, b): #A loop of b times
    result += a  #a added in b times
  return result
```

```python
# Sample function call: func1(5, 3)
import time
start=time.time()
x=print(func1(4,3))
end=time.time()
total_time=(end-start)*1000
print(total_time)
#the function performs multiplication using repeated addition
```

```python
# Exercise 2
def func2(a, b): #Time complexicity is O(b)
  result = 1
  while b > 0: #this fun denotes b should be greater than 0
    result = result * a #this function performs multiplication of a in b times
    b = b - 1
  return result
```

```python
import time
start=time.time()
# Sample function call: func2(2, 4)
x=print(func2(2,4))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```

```python
# Exercise 3
def func3(num): #Time complexicity is O(len(b) log(b)+len(a) len(b)
   result = 0 #loop starts from 0
   while num > 0: #This symbolize num value should be greater than 0
    result = result + num % 10 # This func performs division using repeated addition
    num //= 10 #this func denotes floor division to get a single value of numbers
   return result
```

```python
import time
start=time.time()
# Sample function call: func3(1234)
x=print(func3(1234))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
    #this func performs binary search and gives the output of equal values from both the func values given
```

```python
import time
start=time.time()# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
x=print(func4([8,9,3,4],[7,6,5,0]))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```

```python
# Exercise 5
def func5(n):#Time complexicity is O(n)
    guess = 1
    while guess * guess <= n: #This  function performs multplication till loop matches value
        if guess * guess == n: #if value will be equal than loop will be returned
            return guess
        guess = guess + 1 #this func denotes when while loop still true it will generate the next value
    return None
```

```python
import time
start=time.time()
# Sample function call: func5(16)
x=print(func5(16))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1) #this func denotes the value of given function then loop starts from 0 till its length-1

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
import time
start=time.time()
#Sample function call: func8([3,1,2,7,8,6])
x=print(func8([3,1,2,7,8,6]))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr) #get the length of input arrat
    k = k % arr_len #if K is greater than array normalize its value so that it wont create prblm
    # giving the remainder of the division, which represents the actual shift
    custom_arr = [0] * arr_len
    for i in range(arr_len): # Calculate the new index for the element at current index 'i'
        mystery_index = (i + k) % arr_len
        # Place the element from the original array in its new position
        custom_arr[mystery_index] = arr[i]
    # Return the newly created rotated array
    return custom_arr
```

```python
import time
start=time.time()
#Sample function call: func12([1, 2, 3, 4, 5], 2)
x=print(func12([1,2,3,4,5],2))
end=time.time()
total_time=(end-start)*1000
print(total_time)
```
